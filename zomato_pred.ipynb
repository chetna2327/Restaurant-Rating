**SELECT THE REQUIRED FEATURES/ENTITIES**

```python
df_new = pd.DataFrame(df[["online_order","dish_liked","book_table","votes","location","rest_type","cuisines","approx_cost(for two people)","listed_in(city)","listed_in(type)","rate"]])
```

```python
df_new
```

**CHECK "NA" VALUES IN THE DF**

```python
df_new.isna().sum()
```

**DROP THE NA VALUES**

```python
df_new=df_new.dropna().reset_index (drop =True)
```

**DROP THE REATINGS MENTIONED AS "NEW"**

```python
df_new=df_new[df_new["rate"]!="NEW"]
df_new=df_new[df_new["rate"]!="-"]
df_new=df_new.reset_index (drop =True)
```

```python
df_new
```

**CONVERTING THE RATINGS**

```python
import math
df_new["rate"]=df_new["rate"].apply(lambda x: x.split("/")[0])
df_new["rate"]=df_new["rate"].astype(float)
```

```python
df_new["rate"]
```

```python
df_new["rate"].value_counts()
```

```python
y_df=df_new[["rate"]]
```

**DISHES LIKED**

```python
list_of_dish_liked=list(set([y.lower().strip() for x in df_new["dish_liked"] for y in x.split(",")]))
features_dish_liked=pd.DataFrame([[0]*len(list_of_dish_liked)]*df_new["dish_liked"].shape[0],columns=list_of_dish_liked)
for i,x in enumerate(df_new["dish_liked"]):
    for y in x.split(","):
        yy=y.lower().strip()
        if yy in list_of_dish_liked:
            features_dish_liked.loc[i,yy]=1
```

**CUISINES LIKED**

```python
list_of_cuisines=list(set([y.lower().strip() for x in df_new["cuisines"] for y in x.split(",")]))
features_cuisines=pd.DataFrame([[0]*len(list_of_cuisines)]*df_new["cuisines"].shape[0],columns=list_of_cuisines)
for i,x in enumerate(df_new["cuisines"]):
    for y in x.split(","):
        yy=y.lower().strip()
        if yy in list_of_cuisines:
            features_cuisines.loc[i,yy]=1
```

**NEW DATASET with all features**

```python
df_new_2=pd.concat([features_dish_liked,features_cuisines,df_new], axis=1)
```

**REMOVE THE OLD FEATURES**

```python
df_new_2=df_new_2.drop('cuisines', 1)
df_new_2=df_new_2.drop('dish_liked', 1)
df_new_2=df_new_2.drop('rate', 1)
df_new_2=df_new_2.drop('votes', 1)
```

```python
df_new_2
```

```python
df2=df_new_2
```

**FACTORIZING THE DATA**

```python
df2["online_order"]=df2["online_order"].factorize()[0]
df2["book_table"]=df2["book_table"].factorize()[0]
df2["approx_cost(for two people)"]=df2["approx_cost(for two people)"].apply(lambda x: "".join(x.split(",")))
df2["listed_in(city)"]=df2["listed_in(city)"].str.lower().factorize()[0]
df2["listed_in(type)"]=df2["listed_in(type)"].str.lower().factorize()[0]
df2["location"]=df2["location"].str.lower().factorize()[0]
df2["rest_type"]=df2["rest_type"].str.lower().factorize()[0]
```

```python
df2
```

```python
df4=df2
```

**SPLITING THE DATA INTO TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split  
X_train,X_test,Y_train, Y_test = train_test_split(df4, y_df['rate'], test_size=0.25, random_state=10)
print("Train: ",X_train.shape,Y_train.shape,"Test: ",(X_test.shape,Y_test.shape))
```

**EVAL METRICS FUNCTION**

```python
from sklearn.metrics import mean_absolute_percentage_error
def eval_metrics(rfr,X_test,Y_test):
    print("MAPE: ",100-mean_absolute_percentage_error(Y_test, rfr.predict(X_test)))

    from sklearn.metrics import median_absolute_error
    print("MEDIAN ABSOLUTE: ",100-median_absolute_error(Y_test, rfr.predict(X_test)))


    from sklearn.metrics import r2_score
    print("R2: ",r2_score(Y_test, rfr.predict(X_test))*100)
```

**IMPORT PACKAGES**

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

**LINEAR REGRESSION**

```python
lr=LinearRegression()
lr.fit(X_train, Y_train)
eval_metrics(lr,X_test,Y_test)
```

**RIDGE REGRESSION**

```python
ridg=Ridge()
ridg.fit(X_train, Y_train)
eval_metrics(ridg,X_test,Y_test)
```

**RANDOM FOREST REGRESSION**

```python
rfr = RandomForestRegressor(n_estimators=15, random_state=100)
rfr.fit(X_train, Y_train) 
eval_metrics(rfr,X_test,Y_test)
```

**KNeighborsRegressor**

```python
knr=KNeighborsRegressor()
knr.fit(X_train, Y_train) 
eval_metrics(knr,X_test,Y_test)
```

**SVM**

```python
sv=SVR()
sv.fit(X_train, Y_train) 
eval_metrics(sv,X_test,Y_test)
```
